```python
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score
from dataset import YouTubeDataset
```

```python
splits = [10]
num_classes = len(splits)+1
labels = list(range(num_classes))
dataset = YouTubeDataset(splits)
y_true = list(int(t) for t in dataset.label)

# print("{:.3f} {:.3f} {:.3f} {:.3f}".format(accuracy_score()))
```

```python
majority_list = [0 for _ in range(211)]
print(f"{accuracy_score(y_true, majority_list):.3f}")
print(f"{precision_score(y_true, majority_list, average='macro', labels=labels):.3f}")
print(f"{recall_score(y_true, majority_list, average='macro', labels=labels):.3f}")
print(f"{f1_score(y_true, majority_list, average='macro'):.3f}")
```
